```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/bronze-houseprices.csv')
df.head()
```

```python
df = df.iloc[:,:-2]
df.head()
```

```python
df = df.dropna(subset = ['price', 'title'])
```

```python
df.reset_index(drop=True, inplace=True)
df.index
```

```python
def price_calculator(price):
    import re
    match = re.search(r'\d+\.\d+|\d+', price)  # Handle both decimals & integers
    if not match:
        return None

    num = float(match.group())

    if 'Cr' in price:
        num *= 10000000
    elif 'Lac' in price or 'Lakh' in price:
        num *= 100000

    return num

def price_p_calculator(pps):
    import re
    if not pps:
        return None

    # Extract number (integer or decimal) and remove commas
    match = re.search(r'\d[\d,.]*\.?\d*', pps)
    if not match:
        return None

    num_str = match.group().replace(",", "")
    return float(num_str)
```

```python
df['price'] = df['price'].apply(price_calculator)
```

```python
df = df[~df['price_per_sqft'].isnull() == True]
```

```python
df['price_per_sqft'] = df['price_per_sqft'].astype('string')
```

```python
df['price_per_sqft'] = df['price_per_sqft'].apply(price_p_calculator)
```

```python
# Convert BHK to float
df['bhk_count'] = df['bhk_count'].str.extract(r'(\d+)').astype(float)

# Convert super_area (sqft) to float
df['super_area'] = df['super_area'].str.replace(',', '')  # remove commas
df['super_area'] = df['super_area'].str.extract(r'(\d+\.?\d*)').astype(float)
```

```python
df = df.drop('society_name', axis=1)
```

```python
df['floor'][df['property_type'] == 'Villa'] = 0
```

```python
import re

def extract_floor_number(floor):
    if pd.isna(floor):
        return None
    
    if str(floor).lower().startswith('ground'):
        return 0
    
    # Extract the first number from the floor string
    match = re.search(r'(\d+)', str(floor))
    if match:
        return int(match.group(1))
    
    return None

df['floor'] = df['floor'].apply(extract_floor_number)
```

```python

#Fill missing with mode (most common value)
mode_value = df['floor'].mode()[0]
df['floor'] = df['floor'].fillna(mode_value)
```
